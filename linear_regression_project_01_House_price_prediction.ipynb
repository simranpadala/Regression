```python
y=data ["Price"]
x1=data ["Size"]
```

```python
y
```

```python
x1
```

```python
plt.scatter(x1,y)
plt.xlabel("Size",fontsize=20)
plt.ylabel("Price",fontsize=20)
plt.show()
```

```python
#Training the data with OLS(OLS is created)
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
#c=0.2750
#m=0.0017
```

```python
plt.scatter(x1,y)
yhat=0.0017*x1 + 0.275
fig=plt.plot(x1,yhat,lw=3,c="orange",label="regression line")
plt.xlabel("Size",fontsize=20)
plt.ylabel("Price",fontsize=20)
plt.show()
```
